```python
import pandas as pd
```

```python
import os
```

```python
import music_api


path_filename = os.getcwd() + '/../music_api/dataset.csv'
```

```python
df = pd.read_csv(path_filename)
```

```python
columns_to_keep = [
    'artists', 'track_name', 'popularity',
    'duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness',
    'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
    'valence', 'tempo', 'time_signature']
```

```python
df
```

```python
df = df[columns_to_keep]
```

```python
df = df.sort_values(by='popularity', ascending=False)
```

```python
df = df.drop_duplicates(subset=['track_name'])
```

```python
df
```

```python
df.to_csv(path_filename, index=False)
```

```python
selected_columns = df[['track_name', 'artists']]
```

```python
selected_columns.drop_duplicates()
```

```python
dict_regrouped_by_songs = selected_columns.groupby('track_name')['artists'].apply(list).to_dict()

big_dict = {'songs': dict_regrouped_by_songs}
```

```python
# dump big dicts to json

import json

with open('big_dict.json', 'w') as f:
    json.dump(big_dict, f)
```
